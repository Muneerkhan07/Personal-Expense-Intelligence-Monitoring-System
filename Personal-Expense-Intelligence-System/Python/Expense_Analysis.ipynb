```python
import pandas as pd
import random

dates = pd.date_range(start="2025-01-01", end="2025-12-31")
categories = ["Food", "Travel", "Shopping", "Rent", "Bills", "Entertainment"]

data = []

for date in dates:
    for i in range(2): 
        data.append([
            date.strftime("%Y-%m-%d"),
            random.choice(categories),
            random.randint(200, 3000)
        ])

df = pd.DataFrame(data, columns=["date", "category", "amount"])

df.to_csv("expense_data_2025.csv", index=False)

print("File Created")
df.head()
```
